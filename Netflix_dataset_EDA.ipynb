9. Creamos una función que reciba 3 parámetros, la fila, la columna y el nuevo valor que le asignaremos a la ubicación en cuestión.

```python
def reemplazarValor(df, fila, columna, nuevo):
    try:
        df.at[fila, columna] = nuevo #At en este caso funciona igual que loc
        return True  # Éxito
    except Exception as e:
        pass
```

27. Describimos las columnas con datos numéricos para observar la frecuencia y los datos atípicos

- Histogramas: Para visualizar la frecuencia de cada valor, se utilizan histogramas, que son más adecuados para datos continuos o con muchos valores únicos.

- Diagramas de caja: Se añaden diagramas de caja para identificar visualmente los valores atípicos.

- Muestreo en gráficos de dispersión: Si el DataFrame tiene más de 1000 filas, se toma una muestra aleatoria de 1000 filas para el gráfico de dispersión, mejorando la legibilidad.

```python
def graficar_columnas_numericas(df2, columnas):
    """
    Función para graficar columnas de un DataFrame con mejoras en la visualización.

    Parámetros:
    - df2: DataFrame que contiene los datos.
    - columnas: Lista de nombres de columnas a graficar.

    Genera histogramas, diagramas de caja y gráficos de dispersión muestreados.
    """

    #Comprobamos que las columnas existan en el DataFrame
    for col in columnas:
        if col not in df2.columns:
            print(f"La columna '{col}' no existe en el DataFrame.")
            return

    #Creamos figura con varias subplots
    num_columnas = len(columnas)
    fig, axs = plt.subplots(num_columnas, 3, figsize=(18, 6*num_columnas))

    if num_columnas == 1:
        axs = [axs]  # Aseguramos que axs sea iterable

    for i, col in enumerate(columnas):
        #Histograma
        axs[i][0].hist(df2[col], bins=30, color='skyblue', edgecolor='black')
        axs[i][0].set_title(f'Histograma de {col}')
        axs[i][0].set_xlabel('Valor')
        axs[i][0].set_ylabel('Frecuencia')
        axs[i][0].grid(True, linestyle='--', alpha=0.7)

        #Diagrama de caja
        axs[i][1].boxplot(df2[col], vert=True, patch_artist=True, boxprops=dict(facecolor='skyblue'))
        axs[i][1].set_title(f'Diagrama de caja de {col}')
        axs[i][1].set_ylabel('Valor')
        axs[i][1].grid(True, linestyle='--', alpha=0.7)

        #Gráfico de dispersión muestreado
        if len(df2) > 1000:  #Muestrear si hay demasiados datos
            df_sample = df2.sample(1000, random_state=42)
        else:
            df_sample = df2

        axs[i][2].scatter(range(len(df_sample[col])), df_sample[col], alpha=0.5, color='skyblue')
        axs[i][2].set_title(f'Dispersión (Muestreo) de {col}')
        axs[i][2].set_xlabel('Índice')
        axs[i][2].set_ylabel('Valor')
        axs[i][2].grid(True, linestyle='--', alpha=0.7)

    plt.tight_layout()
    plt.show()
```

```python
graficar_columnas_numericas(df2, ['release_year', 'ratings', 'duration'])
```

28. Describirmos las columnas categóricas

```python
def describir_categoricas(df2):
    """
    Función para describir las columnas categóricas de un DataFrame y generar gráficos de barras.

    Parámetros:
    - df2: DataFrame que contiene los datos.

    Muestra la descripción estadística de las columnas categóricas y genera gráficos de barras con mejoras visuales.
    """

    #Identificamos columnas categóricas (object y category)
    columnas_categoricas = df2.select_dtypes(include=['object', 'category']).columns

    if len(columnas_categoricas) == 0:
        print("No hay columnas categóricas en el DataFrame.")
        return

    #Mostramos descripción estadística
    print("Descripción de las columnas categóricas:\n")
    print(df2[columnas_categoricas].describe())
    print("\n")

    #Generaramos gráficos de barras para cada columna categórica
    num_columnas = len(columnas_categoricas)
    fig, axs = plt.subplots(num_columnas, 1, figsize=(12, 5*num_columnas))

    #Aseguramos que axs sea iterable si hay una sola columna
    if num_columnas == 1:
        axs = [axs]

    for i, columna in enumerate(columnas_categoricas):
        valores = df2[columna].value_counts()

        #Evitamos errores si la columna está vacía
        if valores.empty:
            print(f"La columna '{columna}' no tiene valores.")
            continue

        #Graficamos
        valores.plot(kind='bar', ax=axs[i], color='skyblue', edgecolor='black')
        axs[i].set_title(f'Distribución de {columna}', fontsize=14, fontweight='bold')
        axs[i].set_xlabel('Categoría')
        axs[i].set_ylabel('Frecuencia')
        axs[i].tick_params(axis='x', rotation=45)
        axs[i].grid(axis='y', linestyle='--', alpha=0.7)

    plt.tight_layout()
    plt.show()
```

```python
describir_categoricas(df2[['type', 'rating']])
```

```python
#Creamos el gráfico
fig, ax = plt.subplots(figsize=(20, 9))
fig.set_facecolor('lightgray')

#Graficamos los promedio de ratings por cada 5 años
ratings_por_año.plot(kind='line', marker='o', color='purple', linestyle='-')
plt.title('Evolución de Ratings Promedio Cada 5 Años', fontsize=15, fontweight='bold')
plt.xlabel('Año', fontsize=14)
plt.ylabel('Rating Promedio', fontsize=14)
plt.xticks(rotation=0, fontsize=12)

#Agregamos cuadrículas
plt.grid(True)

plt.show()
```

```python
#Convertimos la columna "date_added" en formato de fecha
programas_df2['date_added'] = pd.to_datetime(programas_df2['date_added'], errors='coerce') # Cuando se establece errors='coerce', significa que si ocurre algún error durante la conversión de los datos, se asignará un valor nulo (NaN) en lugar de generar una excepción o error que detenga la ejecución del programa.

#Usamos groupby para contar películas por año de fecha de agregado
programas_por_año = programas_df2[programas_df2['type'] == 'Programa'].groupby(programas_df2['date_added'].dt.year)['title'].count()

programas_por_año
```

```python
#Creamos el gráfico
fig, ax = plt.subplots(figsize=(12, 7))
fig.set_facecolor('lightgray')

#Graficamos los promedio de ratings por cada 5 años
programas_por_año.plot(kind='line', marker='o', color='purple', linestyle='-')
plt.title('Evolución de la subida de programas a Netflix por año', fontsize=15, fontweight='bold')
plt.xlabel('Año', fontsize=14)
plt.ylabel('Cantidad', fontsize=14)
plt.xticks(rotation=0, fontsize=12)

#Agregamos cuadrículas
plt.grid(True)

plt.show()
```

Análisis 14: Tenemos que hubo un pico que se podría decir que se mantuvo, entre 2019 y 2020, empezando una tendencia hacía este desde el año 2014, como lo dicho anteriormente, es posible que en esos años fue en los que empezó a ganar más popularidad, teniendo cada vez más suscriptores, y por el equipo de Netflix empezó a trabajarle más al proyecto subiendo más contenido.
